```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data =np.loadtxt("Desktop\linerar regression dummy data\data.csv",delimiter=",")
data.shape
```

```python
x=data[:,0].reshape(-1,1)
y=data[:,1].reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
alg1=LinearRegression()
alg1.fit(x_train,y_train)
m=alg1.coef_[0]
c=alg1.intercept_
print(m)
print(c)
x_line=np.arange(20,80,0.1)
y_line=m*x_line+c
```

```python
# graph showing the line made by linear relation y=Wx+R or y=mx+c
plt.plot(x_line,y_line,color="black")
plt.scatter(x_train,y_train)

plt.scatter(x_test,y_test,color="red")
plt.show()
```

```python
# graph between p prediction and y actual
y_pred=alg1.predict(x_test)
plt.scatter(y_test,y_pred)
plt.show()
print(alg1.score(x_test,y_test))
print(alg1.score(x_train,y_train))
```
